## Grafico de barra para demonstrar o fluxo de pedidos A,B e C em semana do mês.

```python
pedidos_por_semana = df.groupby('Week of the month (first week, second, third, fourth or fifth week')[['Order type A', 'Order type B', 'Order type C']].sum()

pedidos_por_semana.plot(kind='bar', figsize=(10, 6))
plt.title('Quantidade de Pedidos por Semana do Mês')
plt.xlabel('Semana do mês')
plt.ylabel('Quantidade de Pedidos')
plt.legend(title='Tipo de Pedido')
plt.show()
```

```python
#Graficos semanas
df['Week of the month (first week, second, third, fourth or fifth week'].value_counts().sort_index().plot(kind='bar', color='skyblue')
plt.title('Distribuição das semanas do mês')
plt.xlabel('Semana do mês')
plt.ylabel('Contagem')
plt.show()
```

```python
#Graficos dias da semana
df['Day of the week (Monday to Friday)'].value_counts().sort_index().plot(kind='bar', color='lightgreen')
plt.title('Distribuição dos dias da semana')
plt.xlabel('Dia da semana')
plt.ylabel('Contagem')
plt.show()
```

```python
#Urgentes e não-urgentes
plt.scatter(df['Non-urgent order'], df['Urgent order'], color='salmon', alpha=0.7)
plt.title('Relação entre pedidos não urgentes e urgentes')
plt.xlabel('Pedidos não urgentes')
plt.ylabel('Pedidos urgentes')
plt.show()
```

```python
#Total de ordens
df['Target (Total orders)'].plot(kind='line', marker='o', linestyle='-', color='purple')
plt.title('Total de pedidos ao longo do tempo')
plt.xlabel('Índice')
plt.ylabel('Total de pedidos')
plt.show()
```

```python
# Vamos verificar se há valores discrepantes usando boxplot = a saída mostra que a coluna para pedidos do setor fiscal
# e Ordens bancárias (1) apresentam valores discrepantes.
f, axes = plt.subplots(3, 6, figsize=(12, 12))

axes = axes.flatten()
columns = ['Non-urgent order', 'Urgent order', 'Order type A', 'Order type B', 'Order type C',
           'Fiscal sector orders', 'Banking orders (1)', 'Orders from the traffic controller sector',
           'Banking orders (2)', 'Banking orders (3)', 'Target (Total orders)']

for col, ax in zip(columns, axes):
    sns.boxplot(y=col, data=df, ax=ax)

for i in range(len(columns), len(axes)):
    f.delaxes(axes[i])

plt.tight_layout()

plt.show()
```

```python
sns.boxplot(x='variable',y='value',data=pd.melt(df[['Order type A','Order type B', 'Order type C']]));
plt.xlabel('Orders');
plt.ylabel('Value');
plt.title('Distribuição dos tipos de pedido');
```

Analisamos abaixo a distribuição dos diferentes tipos de pedido realizados ao longo do recorte temporal. Podemos perceber que os pedidos do tipo B e C são mais elevados do que o pedido do tipo A. No entando, vale frisar que os pedidos B e C também apresentam valores discrepantes ( outliers ), o que gera um valor médio de pedidos maior em relação ao pedido A.

```python
sns.barplot(x='variable', y='value', data=pd.melt(df[['Non-urgent order','Urgent order']]));
plt.xlabel('Orders');
plt.ylabel('Value');
plt.title('Média dos tipos de pedido de acordo com a urgência');
```

Outra forma de identificar a perfomance das vendas é identificar a urgência dos pedidos. Podemos observar que os pedidos não-urgentes possuem uma média maior do que os pedidos urgentes, o que é compreensível até porque a soma total de pedidos não-urgentes é muito maior do que a de urgentes.
