```python
# Box plots
f, axes = plt.subplots(3, 6, figsize=(12, 12))

axes = axes.flatten()
columns = ['Non-urgent order', 'Urgent order', 'Order type A', 'Order type B', 'Order type C',
           'Fiscal sector orders', 'Banking orders (1)', 'Orders from the traffic controller sector',
           'Banking orders (2)', 'Banking orders (3)', 'Target (Total orders)']

for col, ax in zip(columns, axes):
    sns.boxplot(y=col, data=df, ax=ax)

for i in range(len(columns), len(axes)):
    f.delaxes(axes[i])

plt.tight_layout()

plt.show()
```

#### O box plot exibe estatísticas resumidas, como a mediana (linha dentro da caixa), com eles visualizamos a dispersão e a tendência central dos dados, podemos notar valores discrepantes em 'Banking orders (1)' e 'Fiscal sector orders' devido aos valores de 'Outliers' que são valores atípicos.

#### O box plot exibe estatísticas resumidas, como a mediana (linha dentro da caixa) com eles visualizamos a dispersão e a tendência central dos dados. Analisamos a distribuição dos diferentes tipos de pedido realizados no total. Podemos perceber que os pedidos do tipo B e C são mais elevados do que o pedido do tipo A. No entando, vale frisar que os pedidos B e C também apresentam valores discrepantes (outliers), o que gera um valor médio de pedidos maior em relação ao pedido A.

```python
# Gerando histograma
dfe = pd.DataFrame(df)

# Combinando dia da semana (segunda...terça...quarta) com numeração da semana do do ano que pode ser de 1 até 52 semanas
Week_Day = dfe.groupby(['Week of the month (first week, second, third, fourth or fifth week', 'Day of the week (Monday to Friday)']).sum()

# Criando o gráfico de barras
plt.figure(figsize=(14, 10))
bar_width = 0.25
index = np.arange(len(Week_Day))

plt.bar(index - bar_width, Week_Day['Order type A'], bar_width, label='Tipo de pedido A')
plt.bar(index, Week_Day['Order type B'], bar_width, label='Tipo de pedido B')
plt.bar(index + bar_width, Week_Day['Order type C'], bar_width, label='Tipo de pedido C')

# Adicionando linhas tracejadas para cada tipo de pedido
plt.axhline(y=Week_Day['Order type A'].mean(), color='blue', linestyle='--', label='Média Tipo de pedido A')
plt.axhline(y=Week_Day['Order type B'].mean(), color='orange', linestyle='--', label='Média Tipo de pedido B')
plt.axhline(y=Week_Day['Order type C'].mean(), color='green', linestyle='--', label='Média Tipo de pedido C')

plt.xlabel('Dia da Semana, Semana do Mês')
plt.ylabel('Valores')
plt.title('Soma de Pedidos por Dia da Semana e Semana do Mês')
plt.xticks(index, Week_Day.index, rotation=45)
plt.legend()
plt.tight_layout()

plt.show()
```

#### conforme grafico, conseguimos visualizar por dia da semana e semana do mês, durante o periodo '5,2' foi onde houve maior numero de vendas nesta semana

```python
# Correlação entre pedidos urgentes e não urgentes

data = pd.DataFrame(df)

# Calculando a correlação entre pedidos urgentes e não-urgentes
correlation = data['Non-urgent order'].corr(data['Urgent order']) # 

# Criando um gráfico de dispersão
plt.scatter(data['Non-urgent order'], data['Urgent order'])
plt.title('Relação entre pedidos não-urgentes e urgentes')
plt.xlabel('Não urgentes')
plt.ylabel('Urgentes')
plt.show()

print(f"Correlação entre pedidos não-urgentes e urgentes: {correlation}")
```

#### A correlação é um número que varia entre -1 e 1. Quando próximo de 1, indica uma correlação positiva forte, ou seja, à medida que uma variável aumenta, a outra também tende a aumentar. Quando próximo de -1, é uma correlação negativa forte, indicando que conforme uma variável aumenta, a outra tende a diminuir. Se a correlação é próxima de 0, não há uma relação linear clara entre as variáveis. Neste caso há uma correlação positiva, mas não muito forte.  A correlação de Pearson varia de -1 a 1. Em resumo, a correlação de Pearson é uma ferramenta valiosa para entender a relação linear entre variáveis, mas deve ser usada com outras análises para uma compreensão mais completa do relacionamento entre os dados.
